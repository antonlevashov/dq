In this project, we'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations. We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012.

Using visualizations, we can start to explore questions from the dataset like:

- Do students in more popular majors make more money? Using scatter plots
- How many majors are predominantly male? Predominantly female? Using histograms
- Which category of majors have the most students? Using bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

%matplotlib inline
```

```python
recent_grads = pd.read_csv("data/recent-grads.csv")

recent_grads.head()
```

```python
recent_grads.describe()
```

```python
print(recent_grads.iloc[0])
```

```python
# drop rows containing missing values
recent_grads = recent_grads.dropna()
```

## Pandas, Scatter Plots


```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Employed vs. Median')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size and Unemployment_rate')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Full_time and Median')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='ShareWomen and Unemployment_rate')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men and Median')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women and Median')
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

### Pandas, Scatter Matrix Plot


```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
